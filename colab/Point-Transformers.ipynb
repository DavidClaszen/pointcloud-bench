The Point-Transformers repo makes available three implementations. On an L4, one epoch took roughly:

- Menghao: 2-3 minutes
- Hengshuang: 5-6 minutes
- Nico: 4-5 minutes


Some notes:
- This can probably be sped up by using larger batch sizes and learning rates, but that would also change how they were trained originally.
- According to the repo, Hengshuang and Nico don't have fine-tuned hyper-parameters.
- Running a few epochs for each confirms this: Menghao isn't just the fastest, it is also much more accurate.
